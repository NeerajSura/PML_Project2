## Task 1 - Part a : Bayesian Multiple Regression Using Pymc

**Model Definition:**

In linear regression, the model assumes a linear relationship between the independent variables (features) and the dependent variable (target). Mathematically, it can be represented as:  
y = α + β * X + ε  

Where:  
y is the dependent variable (target),  
X is the matrix of independent variables (features),  
α is the intercept (bias term),  
β is the vector of coefficients (weights),  
ε is the error term (residuals).  

**Priors:**  
alpha = pm.Normal('alpha', mu=0, sigma=10):  
Here, we assume a normal prior distribution for the intercept α, with a mean (mu) of 0 and a standard deviation (sigma) of 10.  

beta = pm.Normal('beta', mu=0, sigma=10, shape=X_train.shape[1]):  
Similarly, we assume a normal prior distribution for the coefficients β, with a mean of 0 and a standard deviation of 10. The shape argument is set to the number of features in the training data.  

sigma = pm.HalfNormal('sigma', sigma=1):  
We assume a half-normal prior distribution for the standard deviation of the residuals ε, with a standard deviation of 1. This ensures that the standard deviation is positive.  

**Linear Combination:**  
mu = alpha + pm.math.dot(beta, X_train.T):  
This calculates the mean μ of the dependent variable y based on the linear combination of intercept α, coefficients β, and features X_train.  

**Likelihood:**
p = pm.Normal('p', mu=mu, sigma=sigma, observed=y_train):  
This defines the likelihood of the observed target variable y_train given the mean μ and standard deviation σ. It assumes that the target variable follows a normal distribution with mean μ and standard deviation σ.  

**Sampling:**
trace_regression = pm.sample(1000, tune=1000, return_inferencedata=True):  
This line samples from the posterior distribution of the model parameters using Markov Chain Monte Carlo (MCMC) with 1000 samples and 1000 tuning steps.  

**Posterior Predictive Samples:**  
ppc = pm.sample_posterior_predictive(trace_regression):
This generates samples from the posterior predictive distribution, which represents the distribution of new observations given the observed data and the estimated parameters.  

**Mean Squared Error (MSE):**
mse = ((ppc['p'].mean(axis=0) - y_test) ** 2).mean():  
Finally, this calculates the mean squared error between the mean of the posterior predictive samples and the actual test data y_test, providing a measure of the model's predictive performance.

```python
import pymc as pm

# Define the model
with pm.Model() as regression_model:
    # Priors
    alpha = pm.Normal('alpha', mu=0, sigma=10)
    beta = pm.Normal('beta', mu=0, sigma=10, shape=X_train.shape[1])
    sigma = pm.HalfNormal('sigma', sigma=1)

    # Linear combination
    mu = alpha + pm.math.dot(beta, X_train.T)

    # Likelihood
    Ylikelihood = pm.Normal('p', mu=mu, sigma=sigma, observed=y_train)

    # Sampling
    trace_regression = pm.sample(1000, tune=1000)

# Plotting the trace
pm.plot_trace(trace_regression, compact=True, figsize=(15, 10))
```

```python
import pickle

# Save the trace to a file
with open('trace_regression.pkl', 'wb') as f:
    pickle.dump(trace_regression, f)
```

```python
# Inference
print(pm.summary(trace_regression))
```

```python
pm.plot_autocorr(trace_regression)
```

```python
pm.plots.plot_posterior(trace_regression)
```

### PPD Estimate

```python
# Predictions : Training Samples
with regression_model:
    ppc = pm.sample_posterior_predictive(trace_regression, predictions=True)

# Accessing the predictions from InferenceData object
predictions = ppc["predictions"]

# Accessing the dimensions of the predictions
dimensions = predictions.dims

# Printing the dimensions
print("Dimensions of predictions:", dimensions)

y_pred = np.mean(predictions['p'], axis=(0, 1))
print(type(y_pred))
```

```python
y_pred.shape
```

```python
from sklearn.metrics import r2_score

# Plot predicted vs actual values for test samples, Out of sample errors
plt.figure(figsize=(10, 6))
plt.scatter(y_train, y_pred, alpha=0.5, label='Predictions', color='brown')
plt.plot([y_train.min(), y_train.max()], [y_train.min(), y_train.max()], 'k--', lw=2, label='Identity line')
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')

# Calculate the mean squared error
mse = np.mean((y_pred - y_train) ** 2)
r2 = r2_score(y_train, y_pred)

# Add MSE and R-squared to the plot title
plt.title(f'PPD : Training Samples\nMSE: {mse.item():.2f}, R-squared: {r2:.2f}')

plt.legend()
plt.show()
```

### Point Estimate : Mean

```python
import numpy as np

# Get summary statistics from the trace
summary = pm.summary(trace_regression)

# Extract mean estimates of coefficients and intercept
mean_coefficients = summary.loc[['beta[{}]'.format(i) for i in range(81)], 'mean']
mean_intercept = summary.loc['alpha', 'mean']

# Predict y values for test samples
y_test_pred = np.dot(X_test, mean_coefficients.values) + mean_intercept

```

```python
from sklearn.metrics import r2_score

# Plot predicted vs actual values for test samples, Out of sample errors
plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_test_pred, alpha=0.5, label='Predictions', color='brown')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'k--', lw=2, label='Identity line')
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')

# Calculate the mean squared error
mse = np.mean((y_test_pred - y_test) ** 2)
r2 = r2_score(y_test, y_test_pred)

# Add MSE and R-squared to the plot title
plt.title(f'Point Estimate : Mean - Test Samples\nMSE: {mse.item():.2f}, R-squared: {r2:.2f}')

plt.legend()
plt.show()
```
